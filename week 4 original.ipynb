#                                                         INTRODUCTION

 Visakhapatnam is the executive capital of the state, “Andhra Pradesh” in India. The city has a population of 17.3 lakhs as per the census 2011.Visakhapatnam is lovingly called as “Vizag” by its citizens.
The coastal city got diversified menu in almost all the restaurants. Sea foods being the specialty of the city, the type of dishes they offer cover the tastes right from North India to South India. So, when we are looking out to choose a right location for our restaurant the following aspects of a location are considered:

1.	How accessible is the location?
The Business location should be a free to access for the customers without any special restriction confined to the chosen place.
2.	How visible is the location?
The Business location chosen should be visible to the people so that they can go in whenever there is a desire to eat. (Indirect Brand positioning) 
3.	Who are the competitors?
Are there any competitors? If so how strong are they in that location would we be able to survive such competition in the starting stages.
4.	 What are the places of interest around the location chosen (trending venues)?
If a location has trending venues around it where more people visit those venues daily, there is higher chance they can come to our restaurant.
5.	Is there at least one restaurant in the neighborhood?
As zero restaurants indicates the place has not given some good business in that niche.

# Our assumption is that every restaurant in the city provides Indian food only, so it's not that useful for us to check the "categories" key in each index of the df3's data

```python
#for ids column
Ids=[]
for i in range(len(df3)):
    x=df3[i]["id"]
    Ids.append(x)
#for names column
names=[]
for n in range(len(df3)):
    y=df3[n]["name"]
    names.append(y)
#for location column
Location=[]
for a in range(len(df3)):
    W=df3[a]["location"]
    Location.append(W)
#for latitudes 
Latitude=[]
for q in range(len(Location)):
    X=Location[q]["lat"]
    Latitude.append(X)
# for Longitudes
Longitude =[]
for t in range(len(Location)):
    Y=Location[t]["lng"]
    Longitude.append(Y)
# for address
Address=[]
for j in range(len(Location)):
    Z=Location[j]["formattedAddress"]
    Address.append(Z)
#making the required data frame
required=  pd.DataFrame(list(zip(Ids,names,Latitude,Longitude,Address)),columns=["IDS","NAMES","LATITUDE","LONGITUDE","ADDRESS"])
required

```

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

# Kmeans clustering of visakhapatnam's restaurant data
k=4
vizagrestaurant_clustering=required.drop(["IDS","NAMES","ADDRESS"],1)
kmeans = KMeans(init="k-means++", n_clusters=4, n_init=12).fit(vizagrestaurant_clustering)
kmeans.labels_
required.insert(0, 'Cluster Labels', kmeans.labels_)
    
```

```python
required
```

```python
# create map
import folium
map_clusters =folium.Map(location=[17.7231276,83.3012842],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, address, cluster in zip(required['LATITUDE'], required['LONGITUDE'], required['ADDRESS'], required['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

```python
#Filtering the first cluster
cluster1filt= required["Cluster Labels"].eq(0)

first_cluster=required[cluster1filt]
first_cluster= first_cluster.reset_index(drop=True)
first_cluster
```

```python
# Filtering the second cluster
cluster2filt= required["Cluster Labels"].eq(1)
second_cluster= required[cluster2filt]
second_cluster= second_cluster.reset_index(drop=True)
second_cluster
```

```python
# Filtering the third cluster
cluster3filt= required["Cluster Labels"].eq(2)
third_cluster= required[cluster3filt]
third_cluster= third_cluster.reset_index(drop=True)
third_cluster
```
